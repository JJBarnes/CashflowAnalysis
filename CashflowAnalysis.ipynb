```python
ftse_batches = len(ftse_tickers) / 50
print(ftse_batches)
sp_batches = math.ceil(len(sp500_tickers) / 50)
print(sp_batches)
sp500_tickers = [t.replace(".", "-") for t in sp500_tickers]
```

```python
df = pd.DataFrame()
for x in range(int(sp_batches)):
    x +=1
    print(x)
    print(sp500_tickers[50*x - 50:50*x])
    temp_ticker = Ticker(sp500_tickers[50*x-50:50*x]).cash_flow()
    time.sleep(5)
    df = pd.concat([df, temp_ticker.reset_index()])
    print(df)
```

```python
test = pd.DataFrame
```

```python
# 1. Build historical price df for all tickers
tickers = df['symbol'].unique()
price_dfs = []

for t in tickers:
    #hist = yf.download(t, start=df['asOfDate'].min(), end=df['asOfDate'].max()+pd.Timedelta(days=1), progress=False)
    hist = Ticker(t).history(start=df['asOfDate'].min(), end=df['asOfDate'].max()+pd.Timedelta(days=1))
    hist['symbol'] = t
    hist = hist[['open']]
    hist = hist.reset_index()
    price_dfs.append(hist)
    time.sleep(1)
    print(len(price_dfs))

prices = pd.concat(price_dfs)

```

```python
# 2. Merge with your financial dataframe
prices['date']     = pd.to_datetime(prices['date'])
df_fin = df.sort_values('asOfDate')
df_price = prices.sort_values('date')

test = pd.merge_asof(df_fin, df_price, left_on='asOfDate', right_on = 'date', left_by='symbol', right_by='symbol')
```
